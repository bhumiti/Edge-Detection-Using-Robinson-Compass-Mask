# Robinson Method

## Robinson kernels

```
North= np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
East = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
West= np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
South= np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
North_East= np.array([[-2,-1,0],[-1,0,1],[0,1,2]])
North_West= np.array([[0,1,2],[-1,0,1],[-2,-1,0]])
South_East= np.array([[0,-1,-2],[1,0,-1],[2,1,0]])
South_West= np.array([[2,1,0],[1,0,-1],[0,-1,-2]])
```

## Helper Functions

```
kernels = [North, North_East, East, South_East, 
               South, South_West, West, North_West]
kernel_names = ['North', 'North_East', 'East', 'South_East', 
                    'South', 'South_West', 'West', 'North_West']

def zero_pad(image, k):
  return np.pad(image, k)

def convolve(image, kernel):
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    kernel = np.flipud(np.fliplr(kernel))
    output = np.zeros_like(image)
    padded_image = zero_pad(image, 1)
    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            output[y, x]=(kernel * padded_image[y: y+3, x: x+3]).sum()
    return output
```

# Robinson using in-built for test

```
import scipy
from scipy import ndimage
base='http://www.cs.albany.edu/~xypan/research'
parent_dir = '/gdrive/MyDrive'
os.chdir(parent_dir)
try:
  os.mkdir('Robinson Method In-Built')
except:
  pass
main = os.path.join(parent_dir, 'Robinson Method In-Built')
os.chdir(main)

North= np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
East = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
West= np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
South= np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
North_East= np.array([[-2,-1,0],[-1,0,1],[0,1,2]])
North_West= np.array([[0,1,2],[-1,0,1],[-2,-1,0]])
South_East= np.array([[0,-1,-2],[1,0,-1],[2,1,0]])
South_West= np.array([[2,1,0],[1,0,-1],[0,-1,-2]])
kernels = [North, North_East, East, South_East, 
               South, South_West, West, North_West]
kernel_names = ['North', 'North_East', 'East', 'South_East', 
                    'South', 'South_West', 'West', 'North_West']

def convolve(image, kernel):
  return ndimage.convolve(image,kernel,mode='nearest',cval=0.0)

num = 0
for i in soup.find_all('img'):
  num+=1
  source = base+i['src'][2:]
  image = io.imread(source)
  img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  image_1 = convolve(img, kernels[0])
  image_2 = convolve(img, kernels[1])
  image_3 = convolve(img, kernels[2])
  image_4 = convolve(img, kernels[3])
  image_5 = convolve(img, kernels[4])
  image_6 = convolve(img, kernels[5])
  image_7 = convolve(img, kernels[6])
  image_8 = convolve(img, kernels[7])
  final_frame = cv.vconcat((cv.hconcat((img,image_1,image_2)),cv.hconcat((image_3,image_4,image_5)),cv.hconcat((image_6,image_7,image_8))))
  # cv2_imshow(final_frame)
  cv2.imwrite('{}.png'.format(num), final_frame)
```

# Kirsch Operator

```
import scipy
from scipy import ndimage
base='http://www.cs.albany.edu/~xypan/research'
parent_dir = '/gdrive/MyDrive'
os.chdir(parent_dir)
try:
  os.mkdir('Kirsch')
except:
  pass
main = os.path.join(parent_dir, 'Kirsch')
os.chdir(main)

North1= np.array([[5,5,5],[-3,0,-3],[-3,-3,-3]])
North_West1= np.array([[5,5,-3],[5,0,-3],[-3,-3,-3]])
West1= np.array([[5,-3,-3],[5,0,-3],[5,-3,-3]])
South_West1= np.array([[-3,-3,-3],[5,0,-3],[5,5,-3]])
East1 = -West1
South1= -North1
North_East1= -South_West1
South_East1= -North_West1

kernels = [North1, North_East1, East1, South_East1, 
               South1, South_West1, West1, North_West1]
kernel_names = ['North', 'North_East', 'East', 'South_East', 
                    'South', 'South_West', 'West', 'North_West']

def convolve(image, kernel):
  return ndimage.convolve(image,kernel,mode='nearest',cval=0.0)

num = 0
for i in soup.find_all('img'):
  num+=1
  source = base+i['src'][2:]
  image = io.imread(source)
  img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  image_1 = convolve(img, kernels[0])
  image_2 = convolve(img, kernels[1])
  image_3 = convolve(img, kernels[2])
  image_4 = convolve(img, kernels[3])
  image_5 = convolve(img, kernels[4])
  image_6 = convolve(img, kernels[5])
  image_7 = convolve(img, kernels[6])
  image_8 = convolve(img, kernels[7])
  final_frame = cv.vconcat((cv.hconcat((img,image_1,image_2)),cv.hconcat((image_3,image_4,image_5)),cv.hconcat((image_6,image_7,image_8))))
  cv2_imshow(final_frame)
  # cv2.imwrite('{}.png'.format(num), final_frame)
```

# Prewitt

```
base='http://www.cs.albany.edu/~xypan/research'
parent_dir = '/gdrive/MyDrive'
os.chdir(parent_dir)
try:
  os.mkdir('Prewitt')
except:
  pass
main = os.path.join(parent_dir, 'Prewitt')
os.chdir(main)
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
num = 0
for i in soup.find_all('img'):
  num+=1
  source = base+i['src'][2:]
  image = io.imread(source)
  img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  img_prewittx = cv2.filter2D(img, -1, kernelx)
  img_prewitty = cv2.filter2D(img, -1, kernely)
  final_frame = cv.hconcat((img, img_prewittx, img_prewitty))
  # cv2_imshow(final_frame)
  cv2.imwrite('{}.png'.format(num), final_frame)
```
